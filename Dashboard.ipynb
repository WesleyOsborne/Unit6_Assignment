```python
#Dwelling Types
Census_Data = pd.read_csv('census_data.csv', index_col = "year")
Dwelling_Units = Census_Data.groupby('year').sum()
Dwelling_Unit=Dwelling_Units.drop(["average_house_value", 'shelter_costs_owned', 'shelter_costs_rented'], axis=1)
```

```python
Dwelling_Types = Dwelling_Unit.hvplot.bar(title="Dwelling Types", groupby="year", x="index", rot=90, frame_width=800, frame_height=300).opts(yformatter="%.0f")
```

```python
# Mean Shelter Costs
Census_Data_Mean = Census_Data.groupby('year').mean()
Shelter_Cost = Census_Data_Mean.drop(["average_house_value", "single_detached_house", "apartment_five_storeys_plus", "movable_dwelling",
                                 "semi_detached_house", "row_house", "duplex", "apartment_five_storeys_less", "other_house"], axis=1)
Shelter_Cost_Owned = Shelter_Cost["shelter_costs_owned"]
Shelter_Cost_Owned_Line=Shelter_Cost_Owned.hvplot.line(title="Average Monthly Shelter Cost for Owned Dwellings in Toronto", xlabel="Year", ylabel="Avg Monthly Shelter Costs")
Shelter_Cost_Rented = Shelter_Cost["shelter_costs_rented"]
Shelter_Cost_Rented_Line = Shelter_Cost_Rented.hvplot.line(title="Average Monthly Shelter Cost for Rented Dwellings in Toronto", xlabel="Year", ylabel="Avg Monthly Shelter Costs")
```

```python
#Average House Value
Average_House_Value = Census_Data_Mean["average_house_value"]
Average_House_Value_line=Average_House_Value.hvplot.line(title="Average House Value in Toronto", xlabel="Year", ylabel='Avg. House Value').opts(yformatter="%.0f")
```

```python
plots_as_column = pn.Column("All", Average_House_Value_line, Shelter_Cost_Owned_Line, Shelter_Cost_Rented_Line)
```

```python
Neighbourhood_Ann_House_Value = Census_Data.groupby([Census_Data.index, "neighbourhood"]).mean()
Avg_House_Value = Neighbourhood_Ann_House_Value["average_house_value"]
Avg_House_Value_df=pd.DataFrame(Avg_House_Value).reset_index()
Neighbourhood_Average_House_Value=Avg_House_Value_df.hvplot.line(x="year", y="average_house_value", groupby = "neighbourhood", rot=90).opts(yformatter="%.0f")
```

```python
#Number of Dwelling Types per year 
Census_Data.reset_index()
Census_Dwellings_Data = Census_Data.drop(["average_house_value", "shelter_costs_owned", "shelter_costs_rented"], axis=1)
Census_Dwellings_Data_Bar=Census_Dwellings_Data.hvplot.bar(title="Number of Dwelling Types per Neighbourhood", ylabel="Number of Dwellings", xlabel= "Dwelling Types", groupby="neighbourhood", rot=90, frame_width=800, frame_height=300).opts(yformatter="%.0f")
```

```python
Neighbourhood_Column = pn.Column("Neighbourhood Info", Census_Dwellings_Data_Bar, Neighbourhood_Average_House_Value)
```

```python
# The Top 10 Most Expensive Neighbourhoods
Most_Expensive_Neighbourhoods = Census_Data.groupby('neighbourhood').mean()
Top_Ten = Most_Expensive_Neighbourhoods['average_house_value'].sort_values(ascending=False).head(10)
Top_Ten_Bar = Top_Ten.hvplot.bar(rot=90, ylabel= "Average House Value", title= "Most Expensive Neighbourhoods in Toronto", yformatter="%.0f", frame_width = 600, frame_height = 200)
```

```python
# Neighbourhood Map
Neighbourhood_Locations= Neighbourhood_Locations.groupby('neighbourhood').mean()
Dwellings_Mean = Census_Data.groupby('neighbourhood').mean()
Locations_And_Values = pd.concat([Neighbourhood_Locations, Dwellings_Mean], axis="columns", join="inner")
Locations_And_Values = Locations_And_Values.reset_index()
Locations_And_Values_Map= px.scatter_mapbox(
    Locations_And_Values, 
    lat="lat", 
    lon='lon', 
    size ='average_house_value', 
    color="neighbourhood",
    title= "Average House Values in Toronto")
```

```python
Toronto = pn.Tabs(("Average_Household_Values", Locations_And_Values_Map), 
               ("Yearly Market Analysis", Dwelling_Types),
               ("Shelter Cost vs. House Value", plots_as_column),
               ("Neighbourhood Analysis", Neighbourhood_Column),
               ("Top Expensive Neighbourhoods", Top_Ten_Bar)
              )
```
